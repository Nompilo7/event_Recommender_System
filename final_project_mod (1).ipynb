```python
# Function to classify time of day
def classify_time(time_str):
    hour = int(time_str.split(':')[0])
    if 0 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Late Night'

# Apply function to create 'time_period' column
df['time_period'] = df['Time'].apply(classify_time)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Function to classify day of the week
def classify_day(date):
    day_of_week = date.day_name()
    if day_of_week in ['Saturday', 'Sunday']:
        return 'Weekend'
    else:
        return 'Weekday'

# Apply function to create 'day_of_the_week' column
df['day_of_the_week'] = df['Date'].apply(classify_day)
```

```python
import dill as pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Define the function
def get_recommendations_content(input_content, df, vectorizer):
    # Step 1: Vectorize the input content
    input_vector = vectorizer.transform([input_content])
    
    # Ensure 'tfid_matrix' is the TF-IDF matrix for the DataFrame
    tfid_matrix = vectorizer.fit_transform(df['Content'])
    
    # Step 2: Compute the cosine similarity between the input content and all entries
    input_sim = cosine_similarity(input_vector, tfid_matrix)
    
    # Step 3: Get the similarity scores
    sim_scores = list(enumerate(input_sim[0]))
    
    # Step 4: Sort the entries based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Step 5: Get the indices of the 10 most similar entries
    events_indices = [i[0] for i in sim_scores[:20]]
    

    top_events = df[['Event_id','Event_name', 'Content', 'Location_Province', 'Event_type', 'Month','day','time_period','day_of_the_week']].iloc[events_indices].copy()
    top_events['similarity_score'] = [sim_scores[i][1] for i in range(len(events_indices))]
    return top_events

# Save the function to a pickle file
with open('recommendations_func13.pkl', 'wb') as f:
    pickle.dump(get_recommendations_content, f)

with open('vec13.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
```
